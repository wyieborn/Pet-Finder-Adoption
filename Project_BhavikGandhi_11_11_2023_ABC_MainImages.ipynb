```python
def resize_images(image_files, target_size):
    resized_images = []
    
    for img_path in image_files:
        img = Image.open(img_path)
        img_resized = img.resize(target_size, Image.LANCZOS)
        resized_images.append(img_resized)
    
    return resized_images
```
